```python
import pandas as pd
import time
from bs4 import BeautifulSoup as Soup
import webbrowser
```

```python
movies_data = pd.read_csv('movie_metadata.csv')
```

```python
# write-html.py

f = open('helloworld.html','w')

tic = time.perf_counter()
colours = movies_data.color.value_counts()
toc = time.perf_counter()

html = f"""<html>
<head></head>
<body>&lt;p&gt;&lt;b&gt;¿How many "colour" films and how many "black and white" films are there in the list?&lt;/b&gt;&lt;/p&gt;
&lt;p&gt;Time processed: {toc - tic:0.4f}&lt;/p&gt;&lt;p&gt;{colours}&lt;/p&gt;<div></div></body>
</html>"""

soup = Soup(html)
#f.write(html)
#f.close()

#webbrowser.open_new_tab('helloworld.html')
```

```python
#b = soup.find('/body')
tic = time.perf_counter()
movies_by_dir = movies_data.director_name.value_counts()
toc = time.perf_counter()
newp = soup.new_tag('p')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
    newp = f'<p><b>¿How many films did each director produce?</b></p><p>Time processed:{toc - tic:0.4f}</p><p>{movies_by_dir}</p>'
soup.div.insert_before(newp)
print (soup)
f.write(soup.text)
f.close()

webbrowser.open_new_tab('helloworld.html')
```
